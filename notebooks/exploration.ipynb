```python
import pandas as pd # importar pandas, para manipular los datos como DataFrame, es decir de varios tipos como caracteres, enteros y demas
import matplotlib.pyplot as plt # importar libreria matplotlib que es para vizualizar datos es decir como dashboard

#se carga data limpia

data_patch = "../data/processed/cleaned_products.csv" # se define la ruta
df = pd.read_csv(data_patch) # se carga la data limpia

# analisis y visualizacion

print (" Basic Statistics")
print (df.describe())

plt.hist(df["price"])

plt.title("Distribucion de precios")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()


print("Productos mas caros: ")
highestPrice = df.nlargest(5, 'price')
print (highestPrice)
```
